### Preparing the data
Dividing the data into "attributes" (inputs) and "labels" (outputs).

```python
# Attributes
X = data[["btc_market_cap", "btc_n_transactions", "btc_miners_revenue", "btc_cost_per_transaction", "btc_difficulty", "btc_hash_rate", "btc_cost_per_transaction_percent"]]
X.head(10)
```

```python
# Labels
y = data[["btc_market_price"]]
y.head(10)
```

### Splitting the dataset

Splitting the dataset using scikit-learn module

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print("Dataset splitted successfully")
```

```python
# Comparing the shapes of different training and testing sets
print("Shape of original dataset : ", data.shape)
print("Shape of input training set : ", X_train.shape)
print("Shape of output training set : ", y_train.shape)
print("Shape of input test set : ", X_test.shape)
print("Shape of output test set : ", y_test.shape)
```

### Training the algorithm

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
print("Training complete")
```

### Making Prediction

```python
# testing data
X_test
```

```python
# predicting
y_pred = regressor.predict(X_test)
y_pred
```
